```python
import pandas as pd
from sqlalchemy import create_engine, text
```

```python
engine = create_engine("postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db")
df_cities = pd.DataFrame(engine.connect().execute(text("select replace(name, ' ', '_') as correct_name, * from city;")))
df_cities.head()
```

```python
df_vacation = pd.read_csv('Планирование отпуска.csv', sep = ',')
df_vacation['Город (англ)'] = df_vacation['Город (англ)'].str.replace(' ', '_')
df_vacation
```

```python
df_population = pd.DataFrame(engine.connect().execute(text("select replace(name, ' ', '_') as name, population from city;")))
df_population.head()
```

```python
df_total = df_vacation.merge(right = df_population, left_on = 'Город (англ)', right_on = 'name', how = 'left').fillna(0)
df_total = df_total.drop('name', axis = 1)
df_total
```

```python
df_country_population = df_total[['Страна', 'population']].groupby('Страна').sum()
df_country_population
```
