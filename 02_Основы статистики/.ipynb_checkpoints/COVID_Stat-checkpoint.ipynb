```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

## Сформируем DS

```python
df = pd.read_excel('data_hw_reg.xlsx')
df.head()
```

## Выведем статистику по DS

```python
df.describe()
```

## Вычислим коэффициент корреляции

```python
df['confirmed'].corr(df['deaths'])
```

## Получим уравнение регрессии для прогноза по России.
## Для этого рассчитаем коэффициенты a и b

```python
df_rus = df[df['country'] == 'Russia']
df_us = df[df['country'] == 'US']
```

```python
X = df_rus[['confirmed']]
y = df_rus[['deaths']]
```

```python
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
a = regressor.intercept_
a
```

```python
b = regressor.coef_
b
```

## Построим графики зависимости смертности от заболеваемости по странам и график линейной регрессии для России

```python
plt.title('Динамика зависимости смертности от заболеваемости по странам')
plt.xlabel('Заболеваемость (млн)')
plt.ylabel('Смертность (млн)')
plt.xlim(right = 10 * 10**6)
plt.ylim(top = 3 * 10**5)
plt.plot(df_rus['confirmed'], df_rus['deaths'], label = 'Russia')
plt.plot(df_us['confirmed'], df_us['deaths'], label = 'US')
x_plot = [0.2 * 10**7, 0.8 * 10**7]
y_plot = a + b * x_plot
plt.axline((x_plot[0], y_plot[0][0]), (x_plot[1], y_plot[0][1]), color = 'red', ls = '--', label = 'Линия регрессии')
plt.legend()
plt.show()
```

```python
plt.title('Динамика заболеваемости и смертности в России')
plt.plot(df_rus['date'], df_rus['confirmed'], label = 'Заболеваемость')
plt.plot(df_rus['date'], df_rus['deaths'], label = 'Смертность')
plt.xticks([])
plt.yticks(
            [0,    1 * 10**6, 2 * 10**6, 3 * 10**6, 4 * 10**6, 5 * 10**6, 6 * 10**6],
            ['0', '1 млн.',  '2 млн.',  '3 млн.',  '4 млн.',  '5 млн.',  '6 млн.']            
)
plt.legend()
plt.show()
```

```python
plt.title('Динамика заболеваемости и смертности в США')
plt.plot(df_us['date'], df_us['confirmed'], label = 'Заболеваемость')
plt.plot(df_us['date'], df_us['deaths'], label = 'Смертность')
plt.xticks([])
plt.yticks(
            [ 0,   5 * 10**6, 10 * 10**6, 15 * 10**6, 20 * 10**6, 25 * 10**6, 30 * 10**6, 35 * 10**6],
            ['0', '5 млн.',  '10 млн.',  '15 млн.',  '20 млн.',  '25 млн.',  '30 млн.',  '35 млн.']            
)
plt.legend()
plt.show()
```

## Предскажем смертность при заболеваемости в 25 млн в России

```python
confirmed_pred = 25 * 10**6
death_pred = regressor.predict([[confirmed_pred]])
death_pred
```

## Проверим по уравнению регрессии 

```python
death_pred = a + b * confirmed_pred
death_pred
```

## 2. Это подтверждается графиками динамики заболеваемости и смертности;
## 3. Параметр b уравнения регрессии положителен, что также сведетельствует о прямой зависимости смертности от заболеваемости;
## 4. При уровне заболеваемости в 25 млн. в России ожидаемая смертность составит порядка 582 тыс. человек в день
