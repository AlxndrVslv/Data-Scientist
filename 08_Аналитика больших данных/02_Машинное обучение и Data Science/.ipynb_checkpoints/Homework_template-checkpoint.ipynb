```python
import pandas as pd
```

```python
# Загружаем данные в переменную df из файла homework.csv
```

```python
df = pd.read_csv('homework.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# Построить scatter график по параметрам crim, medv
```

```python
plt.figure(figsize = (12, 6))
plt.scatter(data = df, x = 'crim', y = 'medv')
plt.xlabel('$crim$')
plt.ylabel('$medv$')
plt.show()
```

```python
# Построить scatter график по параметрам rm, medv
```

```python
plt.figure(figsize = (12, 6))
plt.scatter(data = df, x = 'rm', y = 'medv')
plt.xlabel('$rm$')
plt.ylabel('$medv$')

plt.show()
```

```python
fig, axes = plt.subplots(7, 2, figsize = (12, 18))
fig.subplots_adjust(wspace = 0.2, hspace = 0.2)

for i in range(len(df.columns) - 1):
    axes[i // 2, i % 2].scatter(data = df, y = df.columns[i], x = 'medv')
    axes[i // 2, i % 2].set_ylabel(df.columns[i])

axes[i // 2, 0].set_xlabel('medv')
axes[i // 2, 1].set_xlabel('medv')

plt.show()
```

Из графиков видно, что линейная связь атрибута $medv$ прослеживается с атрибутами $rm$, $lstat$

```python
# Сформировать X и y из df, выбрав нужные колонки
```

```python
df.info()
```

```python
df.columns
```

```python
X, y = df[ ['ID', 'crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat'] ], df['medv']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Разбить данные на данные для обучения и проверки
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Создать и обучить модель LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Оценить качество на тестовой выборки
```

```python
mean_squared_error(y_test, model.predict(X_test))
```

Видно, что на всем наборе атрибутов модель линейной регрессии плохо обучилась.
